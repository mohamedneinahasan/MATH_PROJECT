```python
import os
import json

# Path to your dataset file (e.g., "dataset.json")
dataset_path = "/Users/mohamedneinahasan/Downloads/project/datasets /my datatset/scoredataset.json"

# Directory to save individual JSON files
output_dir = "questions_json"
os.makedirs(output_dir, exist_ok=True)

# Load the full dataset
with open(dataset_path, "r") as file:
    dataset = json.load(file)

# Convert each question to an individual JSON file
for idx, item in enumerate(dataset, start=1):
    question_id = f"question_{idx:04d}"  # Format as question_0001, question_0002, etc.
    output_path = os.path.join(output_dir, f"{question_id}.json")
    
    with open(output_path, "w") as output_file:
        json.dump(item, output_file, indent=4)
    
    print(f"Saved: {output_path}")

print("All questions have been converted into individual JSON files.")
```
